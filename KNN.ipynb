## K-nearest neighbors
- A classification algorithm generally used to predict categorical values


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Load the dataset

```python
df = pd.read_csv('SocialNetworkAds.csv')
df
```

### Extracting the independent and dependent variables

```python
# extracting independent variable
x = df.iloc[:, [2,3]].values
x
```

```python
# extracting independent variable
y = df.iloc[:, 4].values
y
```

### Visualize the dataset

```python
sns.heatmap(df.corr())
```

#### we need to convert the categorical values into numeric for plotting 

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

### Feature Scaling

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x_train = sc.fit_transform(x_train)
# y_test = sc.transform(x_test)
```

### Fitting KNN to the training dataset

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric ='minkowski', p =2) 
classifier.fit(x_train, y_train)
```

### Predict the Test Set results

```python
y_pred = classifier.predict(x_test)
y_pred
```

### Evaluation of mthe model through confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
